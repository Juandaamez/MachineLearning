```python
#Author: Juan David Amezquita Nuñez
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Primer punto
dataframe = pd.read_csv("datos_criminalidad.csv")
dataframe.head(10)
```

```python
#Segundo punto
df = dataframe[["Ingreso_Promedio", "Tasa_Criminalidad"]]
df
```

```python
#Tercer punto
plt.plot(df["Ingreso_Promedio"], df["Tasa_Criminalidad"], "c.")
plt.xlabel("Ingreso Promedio")
plt.ylabel("Tasa de Criminalidad")
plt.show()
```

```python
#Cuarto punto
lin_reg = LinearRegression()
lin_reg.fit(df["Ingreso_Promedio"].values.reshape(-1, 1), df["Tasa_Criminalidad"].values)
```

```python
#Quinto punto
theta0 = lin_reg.intercept_
```

```python
#Sexto punto
theta1 = lin_reg.coef_
```

```python
#Septimo punto
X_min_max = np.array([[df["Ingreso_Promedio"].min()], [df["Ingreso_Promedio"].max()]])
y_train_pred = lin_reg.predict(X_min_max)
```

```python
#Octavo punto
plt.plot(X_min_max, y_train_pred, "g-")
plt.plot(df["Ingreso_Promedio"], df["Tasa_Criminalidad"], "c.")
plt.xlabel("Ingreso Promedio")
plt.ylabel("Tasa de Criminalidad")
plt.show()
```

```python
#Noveno punto
#A
# Hacer la predicción
x = 55000
y = theta0 + theta1 * x

print("Tasa de criminalidad estimada para un ingreso promedio de", x, "es", y)

plt.plot(df["Ingreso_Promedio"], df["Tasa_Criminalidad"], "c.")
plt.plot(X_min_max, y_train_pred, "g-")
#Marker
plt.plot(x, y, marker='o', color='r')
plt.xlabel("Ingreso Promedio")
plt.ylabel("Tasa de Criminalidad")
plt.show()
```

```python
#Noveno punto
#B
# Hacer la predicción
x = 35000
y = theta0 + theta1 * x

print("Tasa de criminalidad estimada para un ingreso promedio de", x, "es", y)

plt.plot(df["Ingreso_Promedio"], df["Tasa_Criminalidad"], "c.")
plt.plot(X_min_max, y_train_pred, "g-")
#Marker
plt.plot(x, y, marker='o', color='r')
plt.xlabel("Ingreso Promedio")
plt.ylabel("Tasa de Criminalidad")
plt.show()
```
