```python
#Author: Juan David Amezquita Nu√±ez
import matplotlib.pyplot as plt
import numpy as np
```

```python
#01
concepto = ["Tickets", "Conferencia", "Souvenirs", "Hospedaje", "Alimentaci√≥n", "Transporte", "Onces", "Otros"]
porcentaje = [30, 10, 20, 25, 30, 20, 20, 5]

plt.bar(concepto, porcentaje, color="red")

plt.title("Gastos de viaje de una empresa")
plt.xlabel("Concepto")
plt.ylabel("Porcentaje")

plt.show()
```

```python
#02
respuesta = ["Muy satisfecho", "Moderadamente satisfecho", "Ninguna opini√≥n", "Moderadamente insatisfecho", "Muy insatisfecho"]
cantidad = [29, 55, 5, 20, 9]

plt.barh(respuesta, cantidad, color="blue")

plt.title("Grado de satisfacci√≥n con el seguro m√©dico actual")
plt.xlabel("Cantidad")
plt.ylabel("Respuesta")

plt.show()
```

```python
#03
especie = ["Mam√≠feros", "Aves", "Reptiles", "Anfibios", "Peces"]
en_eeuu = [69, 77, 14, 12, 71]
fuera_eeuu = [251, 175, 64, 8, 11]
```

```python
#03-01
plt.bar(especie, en_eeuu, color="red")
plt.title("N√∫mero de especies salvajes en peligro de extinci√≥n en EE.UU.")
plt.xlabel("Especie")
plt.ylabel("N√∫mero")
plt.show()
```

```python
#03-02
plt.bar(especie, fuera_eeuu, color="orange")
plt.title("N√∫mero de especies salvajes en peligro de extinci√≥n fuera de EE.UU.")
plt.xlabel("Especie")
plt.ylabel("N√∫mero")
plt.show()
```

```python
#03-03
x = np.arange(len(especie))
width = 0.35
plt.bar(x - width/2, en_eeuu, width, color="red", label="En EE.UU.")
plt.bar(x + width/2, fuera_eeuu, width, color="orange", label="Fuera de EE.UU.")
plt.title("Comparaci√≥n del n√∫mero de especies salvajes en peligro de extinci√≥n")
plt.xlabel("Especie")
plt.ylabel("N√∫mero")
plt.xticks(x, especie)
plt.legend()
plt.show()
```

```python
#05
categoria = ["A", "B"]
datos1 = [4, 3]
datos2 = [6, 3]
datos3 = [5, 4]
colores = ["red", "orange", "green"]

y = range(len(categoria))
width = 0.2
for i in range(len(colores)):
    plt.barh(y, eval(f"datos{i+1}"), width, color=colores[i], label=f"Datos {i+1}")
    y = [x + width for x in y]

plt.title("Diagrama de barras horizontal agrupado")
plt.xlabel("Valor")
plt.ylabel("Categor√≠a")

plt.legend()

plt.show()
```

```python
#06
secciones = ["US", "EMEA", "APAC"]
api = [2, 1, 1]
mobile = [5, 3, 2]
web = [8, 3, 2]

ancho = 0.25

pos_api = [i - ancho for i in range(len(secciones))]
pos_mobile = [i for i in range(len(secciones))]
pos_web = [i + ancho for i in range(len(secciones))]

fig, ax = plt.subplots()
ax.bar(pos_api, api, ancho, label="API")
ax.bar(pos_mobile, mobile, ancho, label="Mobile")
ax.bar(pos_web, web, ancho, label="Web")

ax.set_xticks(range(len(secciones)))
ax.set_xticklabels(secciones)
ax.set_ylabel("Cost (sum)")
ax.set_title("Gr√°fico de barras agrupado")

ax.legend()

plt.show()
```

```python
#07
datos = {"A": [30, 75, 70], "B": [20, 82.5, 78], "C": [25, 70, 68], "D": [35, 65, 62]}

fig, ax = plt.subplots()

ancho = 0.2

for i, grupo in enumerate(datos):
    valores = datos[grupo]
    x = grupo
    y = valores[0]
    ax.bar(x, y, ancho, label="Cantidad de Estudiantes")
    y += valores[1]
    ax.bar(x, valores[1], ancho, label="Puntaje Promedio", bottom=valores[0])
    y += valores[2]
    ax.bar(x, valores[2], ancho, label="Puntaje M√°ximo", bottom=y-valores[2])

ax.set_xlabel("Grupos")
ax.set_ylabel("Valores")
ax.set_title("Resultados de un examen de matem√°ticas para cuatro grupos de estudiantes")

ax.legend()

plt.show()
```
